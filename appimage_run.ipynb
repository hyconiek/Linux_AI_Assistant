```python
# Komórka 6: Budowanie aplikacji za pomocą PyInstaller
import os
import subprocess
# Zakładamy, że zmienne globalne jak PROJECT_ROOT_IN_COLAB, VENV_PATH, PYTHON_VENV_EXEC,
# PYINSTALLER_EXEC, PYINSTALLER_APP_NAME, ICON_FILENAME są zdefiniowane w Komórce 4.
# Zmienna pyqt5_dir_path jest ustalana na początku tej komórki.

# Upewnij się, że jesteśmy w katalogu projektu, gdzie jest główny skrypt GUI
if os.getcwd() != PROJECT_ROOT_IN_COLAB:
    print(f"Zmieniam katalog na: {PROJECT_ROOT_IN_COLAB} (był: {os.getcwd()})")
    os.chdir(PROJECT_ROOT_IN_COLAB)
else:
    print(f"Już jestem w katalogu projektu: {PROJECT_ROOT_IN_COLAB}")

# Próba uzyskania ścieżki do katalogu PyQt5 w venv (powtórzenie z poprzedniej wersji dla kompletności)
pyqt5_dir_path = None
print("\nPróba uzyskania ścieżki do katalogu PyQt5 z venv...")
get_pyqt5_path_script = "import PyQt5, os; print(os.path.dirname(PyQt5.__file__))"
try:
    result = subprocess.run([PYTHON_VENV_EXEC, "-c", get_pyqt5_path_script],
                            capture_output=True, text=True, check=True)
    pyqt5_dir_path = result.stdout.strip()
    if os.path.isdir(pyqt5_dir_path):
        print(f"Znaleziona ścieżka do katalogu PyQt5 w venv: {pyqt5_dir_path}")
    else:
        print(f"OSTRZEŻENIE: Uzyskana ścieżka PyQt5 ('{pyqt5_dir_path}') nie jest poprawnym katalogiem.")
        pyqt5_dir_path = None
except subprocess.CalledProcessError as e:
    print(f"BŁĄD podczas próby uzyskania ścieżki PyQt5 z venv: {e.stderr.strip() if e.stderr else 'Brak STDERR'}")
except FileNotFoundError:
    print(f"BŁĄD: Nie znaleziono interpretera Pythona w venv: {PYTHON_VENV_EXEC}")

# --- Definicja ścieżek dla Lorem ipsum.txt ---
# Najpierw spróbuj ścieżki, gdzie jest vendored przez setuptools
lorem_ipsum_source_path_setuptools = os.path.join(VENV_PATH, "lib/python3.11/site-packages/setuptools/_vendor/jaraco/text/Lorem ipsum.txt")
# Potem spróbuj ścieżki, jakby jaraco.text było oddzielnym pakietem
lorem_ipsum_source_path_jaraco = os.path.join(VENV_PATH, "lib/python3.11/site-packages/jaraco/text/Lorem ipsum.txt")

lorem_ipsum_actual_source_path = None
lorem_ipsum_target_dir_in_internal = None

if os.path.exists(lorem_ipsum_source_path_setuptools):
    lorem_ipsum_actual_source_path = lorem_ipsum_source_path_setuptools
    lorem_ipsum_target_dir_in_internal = "setuptools/_vendor/jaraco/text"
    print(f"Znaleziono Lorem ipsum.txt w ścieżce setuptools: {lorem_ipsum_actual_source_path}")
elif os.path.exists(lorem_ipsum_source_path_jaraco):
    lorem_ipsum_actual_source_path = lorem_ipsum_source_path_jaraco
    lorem_ipsum_target_dir_in_internal = "jaraco/text"
    print(f"Znaleziono Lorem ipsum.txt w ścieżce jaraco.text: {lorem_ipsum_actual_source_path}")
else:
    print(f"OSTRZEŻENIE: Nie znaleziono pliku źródłowego Lorem ipsum.txt ani w '{lorem_ipsum_source_path_setuptools}', ani w '{lorem_ipsum_source_path_jaraco}'. Nie zostanie dodany.")

print("\nRozpoczynanie budowania z PyInstaller...")
pyinstaller_command_list = [
    PYINSTALLER_EXEC,
    "--name", PYINSTALLER_APP_NAME,
    "--noconfirm",
    "--onedir",
    "--windowed",
    "--strip",
    # Dane podstawowe
    "--add-data", f"{ICON_FILENAME}:.",      # Ikona aplikacji
    "--add-data", "src:src",               # Twój kod źródłowy z katalogu src
]

# Dodaj cały katalog PyQt5, jeśli został znaleziony
if pyqt5_dir_path and os.path.isdir(pyqt5_dir_path):
    print(f"Dodaję cały katalog PyQt5 ('{pyqt5_dir_path}') przez --add-data jako 'PyQt5'.")
    pyinstaller_command_list.extend(["--add-data", f"{pyqt5_dir_path}:PyQt5"])
else:
    print("OSTRZEŻENIE: Nie znaleziono katalogu PyQt5 lub ścieżka jest niepoprawna. Katalog PyQt5 nie zostanie dodany przez --add-data.")

# Dodaj plik Lorem ipsum.txt, jeśli został znaleziony
if lorem_ipsum_actual_source_path and lorem_ipsum_target_dir_in_internal:
    print(f"Dodaję plik Lorem ipsum: {lorem_ipsum_actual_source_path} do katalogu docelowego: {lorem_ipsum_target_dir_in_internal}")
    # Poprawiona ścieżka docelowa, aby plik trafił do _internal/
pyinstaller_command_list.extend(["--add-data", f"{lorem_ipsum_actual_source_path}:_internal/{lorem_ipsum_target_dir_in_internal}"])

# Wykluczenia modułów
modules_to_exclude = [
    "PyQt5.QtWebSockets", "PyQt5.QtQml", "PyQt5.QtQuick", "PyQt5.QtMultimedia",
    "PyQt5.QtNetworkAuth", "PyQt5.QtBluetooth", "PyQt5.QtNfc", "PyQt5.QtSensors",
    "PyQt5.QtSerialPort", "PyQt5.QtPositioning", "PyQt5.QtLocation", "PyQt5.QtTextToSpeech",
    "PyQt5.QtWebEngineWidgets", "PyQt5.QtWebEngineCore", "PyQt5.QtWebView",
    "PyQt5.QtSql", "PyQt5.QtTest", "PyQt5.QtHelp", "PyQt5.QtDesigner",
    "PyQt5.Qt3DCore", "PyQt5.Qt3DRender", "PyQt5.Qt3DInput", "PyQt5.Qt3DLogic",
    "PyQt5.Qt3DAnimation", "PyQt5.Qt3DExtras", "PyQt5.QtQuick3D"
]
for mod in modules_to_exclude:
    pyinstaller_command_list.extend(["--exclude-module", mod])

# Hidden imports
hidden_imports_list = [
    "PyQt5.sip", "PyQt5.QtCore", "PyQt5.QtGui", "PyQt5.QtWidgets", "PyQt5.QtSvg",
    "PyQt5.QtPrintSupport", # Zostaw jeśli jest szansa na użycie drukowania/zapisu do PDF
    "google.generativeai", "google.ai.generativelanguage", "google.auth", "google.api_core",
    "google.protobuf", "google.type", "google.rpc",
    # "google.longrunning", # Zwykle wciągane przez api_core
    # "google.oauth2",      # Zwykle wciągane przez auth
    "proto", "grpc", "PIL", "pkg_resources", "argparse", "backend_cli"
]
for imp in hidden_imports_list:
    pyinstaller_command_list.extend(["--hidden-import", imp])

# Główny skrypt aplikacji
pyinstaller_command_list.append("linux_ai_assistant_gui.py")

print(f"\nFinalna komenda PyInstaller:\n{' '.join(pyinstaller_command_list)}\n")

try:
    process_pyinstaller = subprocess.run(pyinstaller_command_list, cwd=PROJECT_ROOT_IN_COLAB, check=True, capture_output=True, text=True)
    print("--- PyInstaller STDOUT ---")
    print(process_pyinstaller.stdout)
    if process_pyinstaller.stderr:
        print("--- PyInstaller STDERR ---")
        print(process_pyinstaller.stderr)
    print("\nZakończono budowanie z PyInstaller. Sprawdź katalog 'dist'.")
except subprocess.CalledProcessError as e:
    print("BŁĄD KRYTYCZNY podczas działania PyInstallera!")
    print(f"Return code: {e.returncode}")
    print(f"--- PyInstaller STDOUT (on error) ---\n{e.stdout}")
    print(f"--- PyInstaller STDERR (on error) ---\n{e.stderr}")
    raise
except FileNotFoundError:
    print(f"BŁĄD KRYTYCZNY: Nie znaleziono wykonywalnego PyInstallera: {PYINSTALLER_EXEC}")
    raise

print("\nSprawdzanie zawartości katalogu PyInstallera (dist):")
!find "{PYINSTALLER_DIST_DIR_FULL_PATH}" -maxdepth 2 -ls
print(f"\nW poszukiwaniu libpython w '{PYINSTALLER_DIST_DIR_FULL_PATH}':")
!find "{PYINSTALLER_DIST_DIR_FULL_PATH}" -name "libpython*" -ls
print(f"\nW poszukiwaniu Lorem ipsum.txt w '{os.path.join(PYINSTALLER_DIST_DIR_FULL_PATH, '_internal')}':")
!find "{os.path.join(PYINSTALLER_DIST_DIR_FULL_PATH, '_internal')}" -name "Lorem ipsum.txt" -ls
```

```python
# Komórka 7: Tworzenie struktury AppDir i kopiowanie plików z PyInstallera
import os
import shutil
import subprocess

# Upewnij się, że zmienne globalne są zdefiniowane (powinny być z Komórki 4)
# COLAB_CONTENT_DIR, APPDIR_FULL_PATH, PYINSTALLER_DIST_DIR_FULL_PATH,
# APP_EXECUTABLE_NAME_FROM_PYINSTALLER, APP_EXECUTABLE_NAME_IN_APPDIR_ROOT,
# ICON_DESKTOP_ENTRY_NAME, ICON_APPDIR_TOPLEVEL_FILENAME, ICON_SOURCE_PATH

# Przejdź do katalogu /content, aby AppDir było tworzone na tym samym poziomie
if os.getcwd() != COLAB_CONTENT_DIR:
    print(f"Zmieniam katalog na: {COLAB_CONTENT_DIR} (był: {os.getcwd()})")
    os.chdir(COLAB_CONTENT_DIR)
else:
    print(f"Już jestem w katalogu: {COLAB_CONTENT_DIR}")

# Usuń stary AppDir, jeśli istnieje, i utwórz go na nowo
if os.path.isdir(APPDIR_FULL_PATH):
    print(f"Usuwanie istniejącego katalogu AppDir: {APPDIR_FULL_PATH}...")
    shutil.rmtree(APPDIR_FULL_PATH)
os.makedirs(APPDIR_FULL_PATH, exist_ok=True)
print(f"Utworzono czysty katalog AppDir: {APPDIR_FULL_PATH}")

# --- Kluczowy krok: Kopiowanie TYLKO zawartości z katalogu wyjściowego PyInstallera ---
print(f"\nKopiowanie zawartości z katalogu PyInstallera '{PYINSTALLER_DIST_DIR_FULL_PATH}' do korzenia AppDir '{APPDIR_FULL_PATH}'...")
if os.path.isdir(PYINSTALLER_DIST_DIR_FULL_PATH):
    # Kopiujemy każdy element z katalogu dystrybucyjnego PyInstallera do korzenia AppDir
    for item_name in os.listdir(PYINSTALLER_DIST_DIR_FULL_PATH):
        source_item_path = os.path.join(PYINSTALLER_DIST_DIR_FULL_PATH, item_name)
        destination_item_path = os.path.join(APPDIR_FULL_PATH, item_name)
        if os.path.isdir(source_item_path):
            print(f"  Kopiowanie katalogu: {source_item_path} -> {destination_item_path}")
            shutil.copytree(source_item_path, destination_item_path, dirs_exist_ok=True)
        else:
            print(f"  Kopiowanie pliku: {source_item_path} -> {destination_item_path}")
            shutil.copy2(source_item_path, destination_item_path)
    print("Zawartość katalogu PyInstallera skopiowana.")

    # Zmiana nazwy głównego pliku wykonywalnego w AppDir, jeśli trzeba
    source_exe_in_appdir = os.path.join(APPDIR_FULL_PATH, APP_EXECUTABLE_NAME_FROM_PYINSTALLER)
    target_exe_in_appdir = os.path.join(APPDIR_FULL_PATH, APP_EXECUTABLE_NAME_IN_APPDIR_ROOT)

    if APP_EXECUTABLE_NAME_FROM_PYINSTALLER != APP_EXECUTABLE_NAME_IN_APPDIR_ROOT:
        if os.path.exists(source_exe_in_appdir):
            print(f"Zmieniam nazwę pliku wykonywalnego w AppDir: '{source_exe_in_appdir}' -> '{target_exe_in_appdir}'")
            os.rename(source_exe_in_appdir, target_exe_in_appdir)
        elif not os.path.exists(target_exe_in_appdir):
             print(f"OSTRZEŻENIE: Plik '{source_exe_in_appdir}' nie istnieje w AppDir, a docelowy '{target_exe_in_appdir}' też nie.")

    if not os.path.exists(target_exe_in_appdir):
         print(f"BŁĄD KRYTYCZNY: Główny plik wykonywalny '{target_exe_in_appdir}' nie istnieje w AppDir po kopiowaniu/zmianie nazwy!")
    else:
        print(f"Główny plik wykonywalny poprawnie umieszczony w AppDir: {target_exe_in_appdir}")

        # Poprawka dla libpython (kopiowanie, jeśli PyInstaller spakował bez .1.0)
        print("\nSprawdzanie i przygotowywanie libpython...")
        internal_dir_in_appdir = os.path.join(APPDIR_FULL_PATH, "_internal") # PyInstaller tworzy _internal
        if os.path.isdir(internal_dir_in_appdir): # _internal powinno być skopiowane z dist/
            # Ustal wersję Pythona dynamicznie
            python_major_minor = f"{sys.version_info.major}.{sys.version_info.minor}"
            libpython_packed_name = f"libpython{python_major_minor}.so"
            libpython_expected_name = f"{libpython_packed_name}.1.0"

            libpython_packed_path = os.path.join(internal_dir_in_appdir, libpython_packed_name)
            libpython_expected_path = os.path.join(internal_dir_in_appdir, libpython_expected_name)

            if os.path.exists(libpython_packed_path) and not os.path.lexists(libpython_expected_path):
                print(f"Kopiowanie '{libpython_packed_path}' do '{libpython_expected_path}'...")
                shutil.copy2(libpython_packed_path, libpython_expected_path)
                print(f"Plik libpython skopiowany jako '{libpython_expected_path}'.")
            elif not os.path.exists(libpython_packed_path) and os.path.exists(libpython_expected_path):
                 print(f"Oczekiwany plik '{libpython_expected_path}' już istnieje; spakowany '{libpython_packed_name}' nie jest potrzebny do kopiowania.")
            elif not os.path.exists(libpython_packed_path) and not os.path.exists(libpython_expected_path):
                print(f"OSTRZEŻENIE: Ani '{libpython_packed_path}' ani '{libpython_expected_path}' nie znaleziono w '{internal_dir_in_appdir}'. Sprawdź output PyInstallera.")
            else:
                print(f"Plik libpython ('{libpython_expected_path}' lub '{libpython_packed_name}') wydaje się być poprawnie obecny w _internal.")
        else:
            print(f"OSTRZEŻENIE: Katalog '{internal_dir_in_appdir}' nie istnieje w AppDir. To jest nietypowe po skopiowaniu z PyInstallera. Pomijam operacje na libpython.")
else:
    print(f"BŁĄD KRYTYCZNY: Katalog wyjściowy PyInstallera '{PYINSTALLER_DIST_DIR_FULL_PATH}' nie istnieje. Przebuduj aplikację za pomocą PyInstallera (Komórka 6).")
    raise FileNotFoundError(f"Katalog PyInstallera nie znaleziony: {PYINSTALLER_DIST_DIR_FULL_PATH}")

# --- Tworzenie standardowej struktury AppDir (usr/bin, usr/share, etc.) ---
# Te katalogi są tworzone wewnątrz APPDIR_FULL_PATH
print("\nTworzenie standardowej struktury katalogów w AppDir...")
os.makedirs(os.path.join(APPDIR_FULL_PATH, "usr", "bin"), exist_ok=True)
os.makedirs(os.path.join(APPDIR_FULL_PATH, "usr", "lib"), exist_ok=True)
os.makedirs(os.path.join(APPDIR_FULL_PATH, "usr", "share", "applications"), exist_ok=True)
os.makedirs(os.path.join(APPDIR_FULL_PATH, "usr", "share", "icons", "hicolor", "256x256", "apps"), exist_ok=True)
print("Standardowa struktura AppDir utworzona.")


# --- Tworzenie pliku .desktop ---
print("\nTworzenie pliku .desktop...")
# Plik .desktop na najwyższym poziomie AppDir
DESKTOP_FILE_PATH_IN_APPDIR_ROOT = os.path.join(APPDIR_FULL_PATH, f"{ICON_DESKTOP_ENTRY_NAME}.desktop")
# Plik .desktop w standardowej lokalizacji AppDir/usr/share/applications/
DESKTOP_FILE_PATH_USR = os.path.join(APPDIR_FULL_PATH, "usr", "share", "applications", f"{ICON_DESKTOP_ENTRY_NAME}.desktop")

DESKTOP_FILE_CONTENT = f"""[Desktop Entry]
Version=1.0
Name=Linux AI Assistant
GenericName=AI Command Assistant
Comment=AI Assistant for Linux Commands
Exec={APP_EXECUTABLE_NAME_IN_APPDIR_ROOT}
Icon={ICON_DESKTOP_ENTRY_NAME}
Terminal=false
Type=Application
Categories=Utility;Development;System;
StartupWMClass=Linux AI Assistant
"""
with open(DESKTOP_FILE_PATH_IN_APPDIR_ROOT, "w") as f: f.write(DESKTOP_FILE_CONTENT)
# Kopiowanie pliku .desktop do standardowej lokalizacji w AppDir
shutil.copy2(DESKTOP_FILE_PATH_IN_APPDIR_ROOT, DESKTOP_FILE_PATH_USR)
print(f"Plik .desktop utworzony w '{DESKTOP_FILE_PATH_IN_APPDIR_ROOT}' i skopiowany do '{DESKTOP_FILE_PATH_USR}'.")

# --- Przygotowanie ikon ---
print("\nPrzygotowywanie ikon...")
ICON_DEST_SYSTEM_PATH = os.path.join(APPDIR_FULL_PATH, "usr", "share", "icons", "hicolor", "256x256", "apps", f"{ICON_DESKTOP_ENTRY_NAME}.png")
ICON_DEST_TOPLEVEL_PATH = os.path.join(APPDIR_FULL_PATH, ICON_APPDIR_TOPLEVEL_FILENAME) # np. linux-ai-assistant.png w korzeniu AppDir

if os.path.exists(ICON_SOURCE_PATH):
    shutil.copy2(ICON_SOURCE_PATH, ICON_DEST_SYSTEM_PATH)
    shutil.copy2(ICON_SOURCE_PATH, ICON_DEST_TOPLEVEL_PATH)
    print(f"Ikona skopiowana do '{ICON_DEST_SYSTEM_PATH}' oraz '{ICON_DEST_TOPLEVEL_PATH}'.")

    # Utwórz dowiązanie .DirIcon, jeśli nazwa pliku ikony na najwyższym poziomie różni się od ICON_DESKTOP_ENTRY_NAME.png
    # lub jeśli po prostu chcemy mieć .DirIcon wskazujące na naszą główną ikonę.
    dir_icon_link_path = os.path.join(APPDIR_FULL_PATH, ".DirIcon")
    if not os.path.lexists(dir_icon_link_path):
         # Tworzymy dowiązanie w bieżącym katalogu, więc musimy być w APPDIR_FULL_PATH
        if os.getcwd() != APPDIR_FULL_PATH:
            print(f"Zmieniam katalog na {APPDIR_FULL_PATH} przed utworzeniem .DirIcon")
            os.chdir(APPDIR_FULL_PATH)
        print(f"Tworzenie dowiązania .DirIcon -> {ICON_APPDIR_TOPLEVEL_FILENAME}")
        subprocess.run(["ln", "-sf", ICON_APPDIR_TOPLEVEL_FILENAME, ".DirIcon"], check=False)
    print("Ikony przygotowane.")
else:
    print(f"OSTRZEŻENIE: Plik ikony źródłowej '{ICON_SOURCE_PATH}' nie znaleziony. Ikony nie zostały skopiowane.")

# --- Tworzenie pliku AppRun ---
print("\nTworzenie pliku AppRun...")
APP_RUN_PATH = os.path.join(APPDIR_FULL_PATH, "AppRun")
APP_RUN_CONTENT = f"""#!/bin/bash
APPDIR=$(dirname "$(readlink -f "$0")")
cd "$APPDIR"

# Dodajemy ścieżki do bibliotek i modułów Pythona wewnątrz AppDir
# _internal jest tam, gdzie PyInstaller umieszcza wiele rzeczy, w tym libpython
# Zakładamy, że wszystkie zależności Pythona są w _internal lub w strukturze PyQt5/_internal
export LD_LIBRARY_PATH="$APPDIR/_internal":"$APPDIR/_internal/PyQt5/Qt5/lib${{LD_LIBRARY_PATH:+:$LD_LIBRARY_PATH}}"
export PYTHONPATH="$APPDIR/_internal":"$APPDIR/_internal/src${{PYTHONPATH:+:$PYTHONPATH}}" # src jest wewnątrz _internal po add-data
export QT_PLUGIN_PATH="$APPDIR/_internal/PyQt5/Qt5/plugins${{QT_PLUGIN_PATH:+:$QT_PLUGIN_PATH}}"
export QT_QPA_PLATFORM_PLUGIN_PATH="$APPDIR/_internal/PyQt5/Qt5/plugins/platforms${{QT_QPA_PLATFORM_PLUGIN_PATH:+:$QT_QPA_PLATFORM_PLUGIN_PATH}}"

# Możesz odkomentować poniższe linie do debugowania zmiennych środowiskowych przy uruchamianiu AppImage
# echo "--- AppRun Debug ---" >&2
# echo "APPDIR: $APPDIR" >&2
# echo "LD_LIBRARY_PATH: $LD_LIBRARY_PATH" >&2
# echo "PYTHONPATH: $PYTHONPATH" >&2
# echo "QT_PLUGIN_PATH: $QT_PLUGIN_PATH" >&2
# echo "QT_QPA_PLATFORM_PLUGIN_PATH: $QT_QPA_PLATFORM_PLUGIN_PATH" >&2
# echo "PWD: $(pwd)" >&2
# echo "Uruchamianie: $APPDIR/{APP_EXECUTABLE_NAME_IN_APPDIR_ROOT}" >&2
# echo "--------------------" >&2

exec "$APPDIR/{APP_EXECUTABLE_NAME_IN_APPDIR_ROOT}" "$@"
"""
with open(APP_RUN_PATH, "w") as f: f.write(APP_RUN_CONTENT)
subprocess.run(["chmod", "+x", APP_RUN_PATH], check=True)
print("Plik AppRun utworzony.")

print(f"\n--- Zakończono przygotowywanie struktury {APPDIR_NAME} ---")
print(f"Zawartość korzenia {APPDIR_FULL_PATH}:")
!ls -p "{APPDIR_FULL_PATH}"
if os.path.isdir(os.path.join(APPDIR_FULL_PATH, "_internal")):
    print(f"\nZawartość {os.path.join(APPDIR_FULL_PATH, '_internal')} (pierwsze 15 elementów):")
    !ls -p "{os.path.join(APPDIR_FULL_PATH, '_internal')}" | head -n 15
```
