```python
import os
import subprocess
# import PyQt5 # Usunięto, bo ścieżkę zdobywamy przez subprocess

# --- Definicje zmiennych globalnych (jeśli nie są zdefiniowane wcześniej w notatniku) ---
# Zakładamy, że PROJECT_DIR_NAME jest już zdefiniowane (np. "linux_ai_terminal_assistant")
# Zakładamy, że COLAB_CONTENT_DIR jest już zdefiniowane (np. "/content")

# Jeśli COLAB_CONTENT_DIR lub PROJECT_DIR_NAME nie są zdefiniowane, musisz je tu ustawić:
if 'COLAB_CONTENT_DIR' not in globals():
    COLAB_CONTENT_DIR = "/content"
if 'PROJECT_DIR_NAME' not in globals():
    PROJECT_DIR_NAME = "linux_ai_terminal_assistant" # Ustaw poprawną nazwę

PROJECT_ROOT_IN_COLAB = os.path.join(COLAB_CONTENT_DIR, PROJECT_DIR_NAME)
VENV_PATH = os.path.join(PROJECT_ROOT_IN_COLAB, "venv")
ICON_FILENAME = "app_icon.png" # Upewnij się, że ta zmienna jest zdefiniowana
# --- Koniec definicji zmiennych globalnych ---


# Upewnij się, że jesteś w głównym katalogu projektu, gdzie jest linux_ai_assistant_gui.py
if os.getcwd() != PROJECT_ROOT_IN_COLAB:
    print(f"Zmieniam katalog na: {PROJECT_ROOT_IN_COLAB} (był: {os.getcwd()})")
    os.chdir(PROJECT_ROOT_IN_COLAB)
else:
    print(f"Już jestem w katalogu projektu: {PROJECT_ROOT_IN_COLAB}")

PYINSTALLER_EXEC = os.path.join(VENV_PATH, "bin", "pyinstaller")
PYTHON_VENV_EXEC = os.path.join(VENV_PATH, "bin", "python")

# Ścieżka do katalogu PyQt5 w venv (dla opcji --add-data)
pyqt5_dir_path = None
print("\nPróba uzyskania ścieżki do katalogu PyQt5 z venv...")
get_pyqt5_path_script = "import PyQt5, os; print(os.path.dirname(PyQt5.__file__))"
try:
    result = subprocess.run([PYTHON_VENV_EXEC, "-c", get_pyqt5_path_script],
                            capture_output=True, text=True, check=True)
    pyqt5_dir_path = result.stdout.strip()
    if os.path.isdir(pyqt5_dir_path):
        print(f"Znaleziona ścieżka do katalogu PyQt5 w venv: {pyqt5_dir_path}")
    else:
        print(f"OSTRZEŻENIE: Uzyskana ścieżka PyQt5 nie jest poprawnym katalogiem: {pyqt5_dir_path}")
        pyqt5_dir_path = None
except subprocess.CalledProcessError as e:
    print(f"BŁĄD podczas próby uzyskania ścieżki PyQt5 z venv:")
    print(f"STDERR: {e.stderr.strip() if e.stderr else 'Brak STDERR'}")
    print("Upewnij się, że PyQt5 jest poprawnie zainstalowane w 'venv' (sprawdź komórkę 9).")
except FileNotFoundError:
    print(f"BŁĄD: Nie znaleziono interpretera Pythona w venv: {PYTHON_VENV_EXEC}")
    print("Upewnij się, że komórka tworząca venv (nr 9) wykonała się poprawnie.")


print("\nRozpoczynanie budowania z PyInstaller...")

pyinstaller_command_list = [
    PYINSTALLER_EXEC,
    "--name", "Linux AI Assistant",
    "--noconfirm",
    "--onedir",
    "--windowed",
    "--strip",
    "--add-data", f"{ICON_FILENAME}:.",
    "--add-data", "src:src",
    "--hidden-import=PyQt5",
    "--hidden-import=PyQt5.QtCore",
    "--hidden-import=PyQt5.QtGui",
    "--hidden-import=PyQt5.QtWidgets",
    "--hidden-import=PyQt5.QtSvg",
    "--hidden-import=PyQt5.QtPrintSupport",
    "--hidden-import=PyQt5.sip",
    "--hidden-import=google.generativeai",
    "--hidden-import=google.ai.generativelanguage",
    "--hidden-import=google.auth",
    "--hidden-import=google.api_core",
    "--hidden-import=google.protobuf",
    "--hidden-import=google.type",
    "--hidden-import=google.rpc",
    "--hidden-import=google.longrunning",
    # "--hidden-import=google.iam",
    "--hidden-import=google.oauth2",
    "--hidden-import=proto",
    "--hidden-import=grpc",
    "--hidden-import=PIL",
    "--hidden-import=pkg_resources",
    "--hidden-import=argparse",
    "--hidden-import=backend_cli",
    "linux_ai_assistant_gui.py"
]

if pyqt5_dir_path and os.path.isdir(pyqt5_dir_path):
    print(f"Dodaję cały katalog PyQt5 przez --add-data: {pyqt5_dir_path}")
    # Wstawiamy na pozycję 6, aby było po --windowed, a przed pierwszym --add-data
    # Lista wygląda tak: [PYINSTALLER_EXEC, '--name', 'AppName', '--noconfirm', '--onedir', '--windowed', ...]
    #                               0             1          2            3           4            5
    pyinstaller_command_list.insert(6, f"{pyqt5_dir_path}:PyQt5")
    pyinstaller_command_list.insert(6, "--add-data") # Teraz to będzie na indeksie 6, przesuwając resztę
else:
    print("OSTRZEŻENIE: Nie znaleziono katalogu PyQt5 lub ścieżka jest niepoprawna, nie zostanie dodany przez --add-data jako cały katalog.")

import subprocess # Upewnij się, że jest zaimportowany
print(f"\nUruchamiam PyInstaller z poleceniem:\n{' '.join(pyinstaller_command_list)}\n")
try:
    process_pyinstaller = subprocess.run(pyinstaller_command_list, cwd=PROJECT_ROOT_IN_COLAB, check=True, capture_output=True, text=True)
    print("PyInstaller STDOUT:")
    print(process_pyinstaller.stdout)
    if process_pyinstaller.stderr:
        print("PyInstaller STDERR:")
        print(process_pyinstaller.stderr)
    print("\nZakończono budowanie z PyInstaller. Sprawdź katalog 'dist'.")
except subprocess.CalledProcessError as e:
    print("BŁĄD KRYTYCZNY podczas działania PyInstallera!")
    print(f"Return code: {e.returncode}")
    print("PyInstaller STDOUT (on error):")
    print(e.stdout)
    print("PyInstaller STDERR (on error):")
    print(e.stderr)
    raise
except FileNotFoundError:
    print(f"BŁĄD KRYTYCZNY: Nie znaleziono wykonywalnego PyInstallera: {PYINSTALLER_EXEC}")
    raise
```

```python
import os
import shutil # Do kopiowania plików i katalogów
import subprocess # Do wywoływania poleceń systemowych, jeśli run_command nie jest zdefiniowane

# --- Funkcja pomocnicza do uruchamiania poleceń (jeśli nie zdefiniowana globalnie w notatniku) ---
def run_command(command, shell=False, check=True, cwd=None):
    """Helper function to run shell commands."""
    cmd_str = ' '.join(command) if isinstance(command, list) else command
    effective_cwd = cwd if cwd else os.getcwd()
    print(f"Executing: {cmd_str} (in {effective_cwd})")
    try:
        result = subprocess.run(command, shell=shell, check=check, capture_output=True, text=True, cwd=cwd)
        # Nie drukuj stdout dla każdego polecenia, aby nie zaśmiecać logów, chyba że jest błąd
        if result.returncode != 0 and result.stdout: # Drukuj stdout jeśli był błąd
             print(f"STDOUT (on error for '{cmd_str}'):\n{result.stdout.strip()}")
        if result.stderr:
            print(f"STDERR for '{cmd_str}':\n{result.stderr.strip()}")
        return result
    except subprocess.CalledProcessError as e:
        print(f"Error executing command: {cmd_str}")
        print(f"Return code: {e.returncode}")
        if e.stdout: print(f"STDOUT (on error):\n{e.stdout.strip()}")
        if e.stderr: print(f"STDERR (on error):\n{e.stderr.strip()}")
        if check:
             raise
        return e
    except FileNotFoundError:
        print(f"Error: Command not found during execution of: {cmd_str}")
        if check:
            raise
        return None

# --- Zmienne konfiguracyjne (upewnij się, że są zdefiniowane w poprzednich komórkach Colab) ---
# Zakładamy, że te zmienne są już zdefiniowane i dostępne globalnie w notatniku:
# COLAB_CONTENT_DIR = "/content"
# PROJECT_DIR_NAME = "linux_ai_terminal_assistant"
# PROJECT_ROOT_IN_COLAB = os.path.join(COLAB_CONTENT_DIR, PROJECT_DIR_NAME)
# ICON_FILENAME = "app_icon.png"
# ICON_SOURCE_PATH = os.path.join(PROJECT_ROOT_IN_COLAB, ICON_FILENAME)

APPDIR_NAME = "LinuxAIAssistant.AppDir"
PYINSTALLER_DIST_DIR_NAME = "Linux AI Assistant"
PYINSTALLER_DIST_FULL_PATH = os.path.join(PROJECT_ROOT_IN_COLAB, "dist", PYINSTALLER_DIST_DIR_NAME)
APP_EXECUTABLE_NAME_FROM_PYINSTALLER = "Linux AI Assistant"
APP_EXECUTABLE_NAME_IN_APPDIR_ROOT = "LinuxAIAssistant"
ICON_DESKTOP_ENTRY_NAME = "linux-ai-assistant"
ICON_APPDIR_TOPLEVEL_FILENAME = f"{ICON_DESKTOP_ENTRY_NAME}.png" # Nazwa pliku ikony na najwyższym poziomie AppDir
APPDIR_FULL_PATH = os.path.join(COLAB_CONTENT_DIR, APPDIR_NAME)

# --- Tworzenie i przygotowanie AppDir ---
current_working_directory = os.getcwd()
if current_working_directory != COLAB_CONTENT_DIR:
    print(f"Zmieniam katalog na: {COLAB_CONTENT_DIR} (był: {current_working_directory})")
    os.chdir(COLAB_CONTENT_DIR)
print(f"Bieżący katalog przed operacjami na AppDir: {os.getcwd()}")

if os.path.isdir(APPDIR_FULL_PATH):
    print(f"Usuwanie istniejącego katalogu {APPDIR_FULL_PATH}...")
    shutil.rmtree(APPDIR_FULL_PATH)
os.makedirs(APPDIR_FULL_PATH, exist_ok=True)
os.chdir(APPDIR_FULL_PATH)
print(f"Pracuję w katalogu AppDir: {os.getcwd()} (oczekiwano: {APPDIR_FULL_PATH})")

print(f"Tworzenie podstawowej struktury katalogów w {os.getcwd()}...")
os.makedirs("usr/bin", exist_ok=True)
os.makedirs("usr/lib", exist_ok=True)
os.makedirs("usr/share/applications", exist_ok=True)
os.makedirs(f"usr/share/icons/hicolor/256x256/apps", exist_ok=True)

# --- Krok 1: Kopiowanie całej zawartości z katalogu PyInstallera (dist) do korzenia AppDir ---
print(f"\nKopiowanie zawartości z '{PYINSTALLER_DIST_FULL_PATH}' do '{APPDIR_FULL_PATH}'...")
if os.path.isdir(PYINSTALLER_DIST_FULL_PATH):
    for item in os.listdir(PYINSTALLER_DIST_FULL_PATH):
        s_item = os.path.join(PYINSTALLER_DIST_FULL_PATH, item)
        d_item = os.path.join(APPDIR_FULL_PATH, item)
        if os.path.isdir(s_item):
            shutil.copytree(s_item, d_item, dirs_exist_ok=True)
        else:
            shutil.copy2(s_item, d_item)
    print("Zawartość katalogu PyInstallera skopiowana.")

    source_exe_in_appdir = os.path.join(APPDIR_FULL_PATH, APP_EXECUTABLE_NAME_FROM_PYINSTALLER)
    target_exe_in_appdir = os.path.join(APPDIR_FULL_PATH, APP_EXECUTABLE_NAME_IN_APPDIR_ROOT)

    if APP_EXECUTABLE_NAME_FROM_PYINSTALLER != APP_EXECUTABLE_NAME_IN_APPDIR_ROOT:
        if os.path.exists(source_exe_in_appdir):
            print(f"Zmieniam nazwę '{source_exe_in_appdir}' na '{target_exe_in_appdir}'")
            os.rename(source_exe_in_appdir, target_exe_in_appdir)
        elif not os.path.exists(target_exe_in_appdir):
             print(f"OSTRZEŻENIE: Plik '{source_exe_in_appdir}' nie istnieje, aby zmienić jego nazwę, a docelowy '{target_exe_in_appdir}' też nie istnieje.")

    if not os.path.exists(target_exe_in_appdir):
         print(f"BŁĄD KRYTYCZNY: Główny plik wykonywalny '{target_exe_in_appdir}' nie został poprawnie skopiowany/zmieniony nazwy w AppDir!")
         # Można by tu rzucić wyjątek, aby zatrzymać skrypt
    else:
        print(f"Główny plik wykonywalny w AppDir: {target_exe_in_appdir}")

        # --- POPRAWKA dla libpython: Kopiowanie ---
        print("\nSprawdzanie i przygotowywanie libpython...")
        internal_dir_in_appdir = os.path.join(APPDIR_FULL_PATH, "_internal") # _internal jest teraz w korzeniu AppDir

        if os.path.isdir(internal_dir_in_appdir):
            libpython_packed_name = "libpython3.11.so"
            libpython_expected_name = "libpython3.11.so.1.0"
            libpython_packed_path = os.path.join(internal_dir_in_appdir, libpython_packed_name)
            libpython_expected_path = os.path.join(internal_dir_in_appdir, libpython_expected_name)

            if os.path.exists(libpython_packed_path):
                if not os.path.lexists(libpython_expected_path):
                    print(f"Kopiowanie '{libpython_packed_path}' do '{libpython_expected_path}'...")
                    try:
                        shutil.copy2(libpython_packed_path, libpython_expected_path)
                        print(f"Plik libpython skopiowany jako {libpython_expected_path}")
                    except Exception as e:
                        print(f"BŁĄD: Nie udało się skopiować libpython: {e}")
                else:
                    print(f"Plik lub dowiązanie '{libpython_expected_path}' już istnieje.")
            else:
                print(f"BŁĄD KRYTYCZNY: Spakowany plik '{libpython_packed_path}' nie został znaleziony w '{internal_dir_in_appdir}'!")
        else:
            print(f"OSTRZEŻENIE: Katalog '{internal_dir_in_appdir}' nie istnieje w korzeniu AppDir ('{APPDIR_FULL_PATH}'). Pomijam operacje na libpython.")
        # --- KONIEC POPRAWKI libpython ---
else:
    print(f"BŁĄD KRYTYCZNY: Katalog wyjściowy PyInstallera nie istnieje: {PYINSTALLER_DIST_FULL_PATH}")
    raise FileNotFoundError(f"Katalog PyInstallera nie znaleziony: {PYINSTALLER_DIST_FULL_PATH}")


# --- Krok 2: Tworzenie pliku .desktop ---
print("\nTworzenie pliku .desktop...")
DESKTOP_FILENAME_NO_EXT = ICON_DESKTOP_ENTRY_NAME
# Plik .desktop na najwyższym poziomie AppDir
DESKTOP_FILE_PATH_IN_APPDIR_ROOT = os.path.join(APPDIR_FULL_PATH, f"{DESKTOP_FILENAME_NO_EXT}.desktop")
# Plik .desktop w standardowej lokalizacji AppDir
DESKTOP_FILE_PATH_USR = os.path.join(APPDIR_FULL_PATH, "usr", "share", "applications", f"{DESKTOP_FILENAME_NO_EXT}.desktop")

DESKTOP_FILE_CONTENT = f"""[Desktop Entry]
Version=1.0
Name=Linux AI Assistant
Exec={APP_EXECUTABLE_NAME_IN_APPDIR_ROOT}
Icon={ICON_DESKTOP_ENTRY_NAME}
Terminal=false
Type=Application
Categories=Utility;Development;System;
StartupWMClass=Linux AI Assistant
"""
with open(DESKTOP_FILE_PATH_IN_APPDIR_ROOT, "w") as f: f.write(DESKTOP_FILE_CONTENT)
os.makedirs(os.path.dirname(DESKTOP_FILE_PATH_USR), exist_ok=True)
shutil.copy2(DESKTOP_FILE_PATH_IN_APPDIR_ROOT, DESKTOP_FILE_PATH_USR)
print(f"Plik .desktop utworzony w '{DESKTOP_FILE_PATH_IN_APPDIR_ROOT}' i '{DESKTOP_FILE_PATH_USR}'")

# --- Krok 3: Przygotowanie ikon ---
print("\nPrzygotowywanie ikon...")
ICON_DEST_SYSTEM_PATH = os.path.join(APPDIR_FULL_PATH, "usr", "share", "icons", "hicolor", "256x256", "apps", f"{ICON_DESKTOP_ENTRY_NAME}.png")
ICON_DEST_TOPLEVEL_PATH = os.path.join(APPDIR_FULL_PATH, ICON_APPDIR_TOPLEVEL_FILENAME)

os.makedirs(os.path.dirname(ICON_DEST_SYSTEM_PATH), exist_ok=True)

if os.path.exists(ICON_SOURCE_PATH):
    shutil.copy2(ICON_SOURCE_PATH, ICON_DEST_SYSTEM_PATH)
    shutil.copy2(ICON_SOURCE_PATH, ICON_DEST_TOPLEVEL_PATH)

    link_for_desktop_icon_name = os.path.join(APPDIR_FULL_PATH, f"{ICON_DESKTOP_ENTRY_NAME}.png")
    if ICON_APPDIR_TOPLEVEL_FILENAME != f"{ICON_DESKTOP_ENTRY_NAME}.png":
        if os.path.exists(ICON_DEST_TOPLEVEL_PATH) and not os.path.lexists(link_for_desktop_icon_name):
            print(f"Tworzenie dowiązania ikony w korzeniu AppDir: {link_for_desktop_icon_name} -> {ICON_APPDIR_TOPLEVEL_FILENAME}")
            # run_command tworzy dowiązanie w bieżącym katalogu, jeśli cwd nie jest podane
            # upewnijmy się, że jesteśmy w APPDIR_FULL_PATH
            if os.getcwd() != APPDIR_FULL_PATH: os.chdir(APPDIR_FULL_PATH)
            subprocess.run(["ln", "-sf", ICON_APPDIR_TOPLEVEL_FILENAME, f"{ICON_DESKTOP_ENTRY_NAME}.png"], check=False)
else:
    print(f"OSTRZEŻENIE: Plik ikony źródłowej nie znaleziony: {ICON_SOURCE_PATH}")

# --- Krok 4: Tworzenie AppRun ---
print("\nTworzenie pliku AppRun...")
APP_RUN_PATH = os.path.join(APPDIR_FULL_PATH, "AppRun")
APP_RUN_CONTENT = f"""#!/bin/bash
APPDIR=$(dirname "$(readlink -f "$0")")
cd "$APPDIR"

# Dodajemy ścieżki do bibliotek i modułów Pythona wewnątrz AppDir
# _internal jest tam, gdzie PyInstaller umieszcza wiele rzeczy, w tym libpython
# Katalog główny AppDir zawiera resztę spakowanych plików Pythona
export LD_LIBRARY_PATH="$APPDIR/_internal":"$APPDIR/usr/lib":"$APPDIR/lib${{LD_LIBRARY_PATH:+:$LD_LIBRARY_PATH}}"
export PYTHONPATH="$APPDIR:$APPDIR/_internal${{PYTHONPATH:+:$PYTHONPATH}}"
# QT_PLUGIN_PATH może być potrzebne dla wtyczek Qt
export QT_PLUGIN_PATH="$APPDIR/usr/plugins":"$APPDIR/plugins":"$APPDIR/_internal/PyQt5/Qt5/plugins${{QT_PLUGIN_PATH:+:$QT_PLUGIN_PATH}}"

echo "--- AppRun Environment ---" >&2
echo "APPDIR=$APPDIR" >&2
echo "LD_LIBRARY_PATH=$LD_LIBRARY_PATH" >&2
echo "PYTHONPATH=$PYTHONPATH" >&2
echo "QT_PLUGIN_PATH=$QT_PLUGIN_PATH" >&2
echo "Wykonywanie: ./{APP_EXECUTABLE_NAME_IN_APPDIR_ROOT}" >&2
echo "--------------------------" >&2

exec ./{APP_EXECUTABLE_NAME_IN_APPDIR_ROOT} "$@"
"""
with open(APP_RUN_PATH, "w") as f: f.write(APP_RUN_CONTENT)
subprocess.run(["chmod", "+x", APP_RUN_PATH], check=True) # Użyj subprocess.run
print("Plik AppRun utworzony z rozszerzonymi ścieżkami.")

print("\nStruktura utworzonego AppDir (wybrane elementy):")
print(f"\nPliki w korzeniu {APPDIR_FULL_PATH}:")
subprocess.run(["ls", "-p", APPDIR_FULL_PATH], check=False)
internal_path_to_check = os.path.join(APPDIR_FULL_PATH, "_internal")
if os.path.isdir(internal_path_to_check):
    print(f"\nZawartość {internal_path_to_check} (w poszukiwaniu libpython):")
    subprocess.run(["find", internal_path_to_check, "-name", "libpython*"], check=False)
else:
    print(f"\nKatalog {internal_path_to_check} nie istnieje.")

print(f"\nZawartość {os.path.join(APPDIR_FULL_PATH, 'usr/share/applications')}:")
subprocess.run(["ls", "-l", os.path.join(APPDIR_FULL_PATH, "usr/share/applications")], check=False)
print(f"\nZawartość {os.path.join(APPDIR_FULL_PATH, 'usr/share/icons/hicolor/256x256/apps')}:")
subprocess.run(["ls", "-l", os.path.join(APPDIR_FULL_PATH, "usr/share/icons/hicolor/256x256/apps")], check=False)

print(f"\n--- Zakończono przygotowywanie {APPDIR_NAME} ---")
```
